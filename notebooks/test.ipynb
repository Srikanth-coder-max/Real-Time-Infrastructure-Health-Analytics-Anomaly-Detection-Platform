```python
import sqlite3
import pandas as pd
from pathlib import Path

db_path = Path("..") / "performance_logs.db"
conn = sqlite3.connect(db_path)
conn.execute("""
    CREATE TABLE IF NOT EXISTS system_logs (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        timestamp TEXT,
        cpu REAL,
        memory REAL,
        is_anomaly INTEGER
    )
""")

df = pd.read_sql("SELECT * FROM system_logs ORDER BY id DESC LIMIT 10", conn)
df
```
