```python
df["CPU_Usage"] = df["CPU_Usage"].clip(0, 100)
df["Memory_Usage"] = df["Memory_Usage"].clip(0, 100)
```

```python
df["cpu_scaled"] = df["CPU_Usage"] / 100
df["memory_scaled"] = df["Memory_Usage"] / 100
```

```python
clean_df = df[["timeStamp", "cpu_scaled", "memory_scaled"]]
```

```python
import os

print(os.listdir("data/processed"))
```

```python
Path("data/processed").mkdir(parents=True, exist_ok=True)
clean_df.to_csv("data/processed/clean_data.csv", index=False)
```
